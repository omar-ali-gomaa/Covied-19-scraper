```python
for idx , d in enumerate (response['rawData']):
   df.loc[idx, 'country_name'] = d['Country_Region']
   df.loc[idx, 'confirmed'] = d['Confirmed']
   df.loc[idx, 'deaths'] = d['Deaths']
```
